### torch.nn.functional 很多层和函数在这里都会见到

torch.nn.functional中有很多功能，后续会常用的。那什么时候使用nn.Module，什么时候使用nn.functional呢？一般情况下，如果模型有可学习的参数，最好用nn.Module，其他情况nn.functional相对更简单一些

```python
import torch.nn.functional as F

loss_func = F.cross_entropy

# 计算预测值
def model(xb):
    return xb.mm(weights) + bias
```

```python
bs = 64
xb = x_train[0:bs]  # a mini-batch from x
yb = y_train[0:bs]  # yb为真实值
weights = torch.randn([784, 10], dtype = torch.float,  requires_grad = True) # requires_grad = True表示需要更新
bs = 64
bias = torch.zeros(10, requires_grad=True)

print(loss_func(model(xb), yb))
```

### 创建一个model来更简化代码

- 必须继承nn.Module且在其构造函数中需调用nn.Module的构造函数
- 无需写反向传播函数，nn.Module能够利用autograd自动实现反向传播
- Module中的可学习参数可以通过named_parameters()或者parameters()返回迭代器

```python
from torch import nn

class Mnist_NN(nn.Module):
    def __init__(self):
        super().__init__()
        # 类初始化时，定义好隐藏层和输出层的 神经元大小 和 权重矩阵大小
        self.hidden1 = nn.Linear(784, 128)
        self.hidden2 = nn.Linear(128, 256)
        self.out  = nn.Linear(256, 10)
        # 使用dropout，防止过拟合，每次杀死一半的数据。（按照多少百分比去杀死数据）。按理说每一层的全连接都要加上dropout的，卷积层一般不加这个
        self.dropout = nn.Dropout(0.5)

    # 前向传播（使用relu激活函数）
    def forward(self, x):
        x = F.relu(self.hidden1(x))
        self.dropout(x)
        x = F.relu(self.hidden2(x))
        self.dropout(x)
        x = self.out(x)
        return x
        
```

```python
net = Mnist_NN()
print(net)
```

可以打印我们定义好的每一层的 名字里的权重和偏置项（我们在定义类时，没有做权重和偏置，都是模型给我们做好了的）

```python
for name, parameter in net.named_parameters():
    print(name, parameter,parameter.size())
```

### 使用TensorDataset和DataLoader来简化（构建数据集）

```python
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

# 先把数据集转换为 TensorDataset格式的数据
train_ds = TensorDataset(x_train, y_train)
# DataLoader现在先理解它是一个 打包的，给我们完整的数据集打包的。batch_size就是每次打包的大小，一批一批数据训练，一批就是一个batch
# shuffle就是一个洗牌的操作，打乱数据的。
train_dl = DataLoader(train_ds, batch_size=bs, shuffle=True)

valid_ds = TensorDataset(x_valid, y_valid)
valid_dl = DataLoader(valid_ds, batch_size=bs * 2)
```

```python
def get_data(train_ds, valid_ds, bs):
    return (
        DataLoader(train_ds, batch_size=bs, shuffle=True),
        DataLoader(valid_ds, batch_size=bs * 2),
    )
```

- 一般在训练模型时加上model.train()，这样会正常使用Batch Normalization和 Dropout，会正常更新每一层的权重和偏置
- 测试的时候一般选择model.eval()，这样就不会使用Batch Normalization和 Dropout，不会更新权重和偏置

```python
import numpy as np

# 上面完成了model的构建 以及 数据集的打包。现在需要把它们两个揉在一下，做train和validation的操作了
'''
steps：数据要迭代多少次
opt：优化器（SGD:梯度下降等等）
'''
def fit(steps, model, loss_func, opt, train_dl, valid_dl):
    for step in range(steps):
        model.train()
        for xb, yb in train_dl:
            losses, nums = loss_batch(model, loss_func, xb, yb, opt)
            print(losses, nums)

        model.eval()
        # no_grad没有梯度，就是不更新w和d
        with torch.no_grad():
            # 这一步把每一个元组中对应的损失和batch分开，分为一个损失元组和一个batch元组，分开完之后它们还是一一对应的
            losses, nums = zip(
                *[loss_batch(model, loss_func, xb, yb) for xb, yb in valid_dl]
            )
        # 计算validation验证集结果；计算平均损失
        # np.sum(np.multiply(losses, nums))：将losses，nums两个元组对应位置相乘再相加
        # np.sum(nums)：样本总和
        val_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)
        print('当前step:'+str(step), '验证集损失：'+str(val_loss))
```

```python
from torch import optim
# 实例化模型 和 初始化优化器
def get_model():
    model = Mnist_NN()
    '''
     SGD: 使用梯度下降的优化器。
     lr: 学习率，一开始要小一点
     有两个参数：第一个参数表示你要更新哪些参数（每一层的w和b都更新）
    '''
    return model, optim.SGD(model.parameters(), lr=0.001)
```

```python
def loss_batch(model, loss_func, xb, yb, opt=None):
    # 计算损失
    loss = loss_func(model(xb), yb)

    # 更新w和b
    if opt is not None:
        # 反向传播
        loss.backward()
        # step相当于实际去执行的更新，向梯度下降的方向，走学习率那么大小
        opt.step()
        # 经过多次迭代，按最新的一次数据迭代，之前的数据抛弃（所以更新一次梯度之前，要置零）
        opt.zero_grad()
    # 返回当前 损失和batch值（之前将训练集打成多个包，每一个包就是一个batch，计算损失时，一个batch对应一个损失值）
    return loss.item(), len(xb)
```

### 三行搞定！

```python
train_dl, valid_dl = get_data(train_ds, valid_ds, bs)
model, opt = get_model()
fit(25, model, loss_func, opt, train_dl, valid_dl)
```

#### 练习
- 1.将优化器由SGD改为Adam
- 2.更改网络层数和神经元个数观察效果
- 3.计算当前模型的准确率等于多少

```python
corrent = 0
total = 0
for xb, yb in valid_dl:
    outputs = model(xb)
    _, predicted = torch.max(outputs.data, 1) # 返回最大的值和索引。1代表从同一个样本中找最大值，0代表从不同样本中找最大值。
    total += yb.size(0)
    corrent += (predicted == yb).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
100 * corrent) / total)
```
